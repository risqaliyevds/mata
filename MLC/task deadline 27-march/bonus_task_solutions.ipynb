```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

import seaborn as sns
```

# Task 1

```python
x_a_train = [-2, -2, -2, -2, -1, -1, -1, -1, -0.8, -0.8, -0.8, -0.8, -0.4, -0.4, -0.4, -0.4,
            0.4, 0.4, 0.8, 0.8, 0.8, 1,   1.5, 1.5]
y_a_train = [-1.5, -1, -0.4, -0.2, -1.5, -1, -0.4, -0.2, -1.5, -1, -0.4, -0.2, -1.5, -1, -0.4, -0.2, 
            1,   0.8, 0.8, 1,   1.5, 0.2, 0.2, 1.5]
```

```python
x_a_test = [0.2, 0.2, 0.8, 1,   1, 1,   1.5, 1.5]
y_a_test = [0.2, 1.5, 0.2, 0.8, 1, 1.5, 0.8, 1]
```

```python
x_b_train = [-1.5, -1.5, -1.5, -1,  -1, -1, -1, -0.8, -0.8, -0.8, -0.2, -0.2,
            0.5, 0.5, 0.5, 0.5, 0.8, 0.8, 0.8, 0.8, 1, 1, 1, 1, 1.5, 1.5, 1.5]
y_b_train = [0.4,   0.8,    1, 0.2, 0.4, 1, 1.5, 0.2,  0.8,  1.5, 0.5,  1.5,
            -1.5, -1, -0.8, -0.4, -1.5, -1, -0.8, -0.4, -1.5, -1, -0.8, -0.4, -1.5, -0.8, -0.4, ]
```

```python
x_b_test = [-1.5, -0.8, -0.2, -0.2, 1.5]
y_b_test = [1.5,    1,    1,   0.8, -1]
```

```python
plt.xlim(right=3)
plt.xlim(left=-3)

plt.ylim(top=3)
plt.ylim(bottom=-3)

plt.plot(x_a_train, y_a_train, 'bo', x_a_test, y_a_test, 'b*', 
         x_b_train, y_b_train, 'ro', x_b_test, y_b_test, 'r*')
```

```python
X_train = []
Y_train = []
for i, j in zip(x_a_train, y_a_train):   # let A be 0 and B to be 1
    X_train.append((i, j))
    Y_train.append(0)

for i, j in zip(x_b_train, y_b_train):   # let A be 0 and B to be 1
    X_train.append((i, j))
    Y_train.append(1)
```

```python
X_test = []
Y_test = []
for i, j in zip(x_a_test, y_a_test):   # let A be 0 and B to be 1
    X_test.append((i, j))
    Y_test.append(0)

for i, j in zip(x_b_test, y_b_test):   # let A be 0 and B to be 1
    X_test.append((i, j))
    Y_test.append(1)
```

```python
model = LogisticRegression().fit(X_train, Y_train)
```

```python
model.score(X_test, Y_test)
```

# Solution

```python
for index, value in enumerate(x_a_train):
    x_a_train[index] = value * y_a_train[index]
    
for index, value in enumerate(x_b_train):
    x_b_train[index] = value * y_b_train[index]
    
for index, value in enumerate(x_a_test):
    x_a_test[index] = value * y_a_test[index]
    
for index, value in enumerate(x_b_test):
    x_b_test[index] = value * y_b_test[index]
```

```python
plt.plot(x_a_train, y_a_train, 'bo', x_a_test, y_a_test, 'b*', 
         x_b_train, y_b_train, 'ro', x_b_test, y_b_test, 'r*')
```

```python
X_train = []
Y_train = []
for i, j in zip(x_a_train, y_a_train):   # let A be 0 and B to be 1
    X_train.append((i, j))
    Y_train.append(0)

for i, j in zip(x_b_train, y_b_train):   # let A be 0 and B to be 1
    X_train.append((i, j))
    Y_train.append(1)
```

```python
X_test = []
Y_test = []
for i, j in zip(x_a_test, y_a_test):   # let A be 0 and B to be 1
    X_test.append((i, j))
    Y_test.append(0)

for i, j in zip(x_b_test, y_b_test):   # let A be 0 and B to be 1
    X_test.append((i, j))
    Y_test.append(1)
```

```python
model = LogisticRegression().fit(X_train, Y_train)
```

```python
model.score(X_test, Y_test)
```

# Solution

```python
sns.pairplot(
    data = df,
    hue = 'label',
    corner = True
    ).fig.suptitle(
    'Feature dependencies',
    y = 1.08
);
```

# Pay attention to the circle in f_1 vs. f_2 !

## It is perfect circle ! Formula of circle: (x-a)^2 - (y-b)^2 = r^2

Let's take the radius as a feature:

```python
df['sq_r'] = (df['features_1'] ** 2 + df['features_2'] ** 2)
```

```python
sns.pairplot(
    data = df,
    hue = 'label',
    corner = True
    ).fig.suptitle(
    'Feature dependencies',
    y = 1.08
);
```

```python
X = df['sq_r'].values.reshape(-1, 1)
y = df['label']
X_train, X_test, y_train, y_test = train_test_split(
X, y, test_size=0.20, random_state=42)
model = LogisticRegression().fit(X_train, y_train)
model.score(X_test, y_test)
```

# Task 3 with sol.

```python
path = "https://raw.githubusercontent.com/a-milenkin/Datasetes_for_Piplines/main/interview/complex_classification.csv"
df = pd.read_csv(path, index_col=0)
df
```

```python
sns.pairplot(
    data = df,
    hue = 'label',
    corner = True
    ).fig.suptitle(
    'Feature dependencies',
    y = 1.08
);
```

```python
df_1 = df[df["features_1"] > 5]
df_1['sq_r'] = ( (df_1['features_1'] - 10) ** 2 + (df_1['features_2'] - 10) ** 2)

df_2 = df[df["features_1"] < 5]
df_2['sq_r'] = (df_2['features_1'] ** 2 + df_2['features_2'] ** 2)
```

```python
sns.pairplot(
    data = df_1,
    hue = 'label',
    corner = True
    ).fig.suptitle(
    'Feature dependencies',
    y = 1.08
);
```

```python
sns.pairplot(
    data = df_2,
    hue = 'label',
    corner = True
    ).fig.suptitle(
    'Feature dependencies',
    y = 1.08
);
```

```python
df = pd.concat([df_1, df_2])
df
```

```python
sns.pairplot(
    data = df,
    hue = 'label',
    corner = True
    ).fig.suptitle(
    'Feature dependencies',
    y = 1.08
);
```

```python
X = df['sq_r'].values.reshape(-1, 1)
y = df['label']
X_train, X_test, y_train, y_test = train_test_split(
X, y, test_size=0.20, random_state=42)
model = LogisticRegression().fit(X_train, y_train)
model.score(X_test, y_test)
```
